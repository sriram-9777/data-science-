```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
```

```python
dataset
```

```python
print(dataset['DESCR'])
```

```python
df_diabetes = pd.DataFrame(dataset.data, columns=['age', 'sex', 'bmi', 'bp', 's1', 's2', 's3', 's4', 's5', 's6'])
df_diabetes.head()
```

```python
df_diabetes.shape
```

```python
X = df_diabetes
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train.shape, y_train.shape
```

```python
X_train.head()
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(X_train.corr(), annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor.fit(X_train, y_train)
```

```python
param = {
    'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
param
```

```python
from sklearn.model_selection import GridSearchCV
regressor = DecisionTreeRegressor()
grid = GridSearchCV(regressor, param, cv=5, scoring='neg_mean_squared_error')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit (X_train, y_train)
```

```python
grid.best_params_
```

```python
y_train_pred = grid.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print(r2_score(y_train, y_train_pred))
print(mean_squared_error(y_train, y_train_pred))
print(mean_absolute_error(y_train, y_train_pred))
```

```python
model_with_best=DecisionTreeRegressor(criterion= 'friedman_mse',max_depth= 5, max_features = 'sqrt',splitter= 'random')
```

```python
model_with_best.fit(X_train, y_train)
```

```python
y_train_pred = model_with_best.predict(X_train)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

print(r2_score(y_train, y_train_pred))
print(mean_squared_error(y_train, y_train_pred))
print(mean_absolute_error(y_train, y_train_pred))
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model_with_best,filled=True)
```
