```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
iris.data
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.head()
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(iris_data)
```

```python
x_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca_scalled = pca.fit_transform(x_scaled)
```

```python
pca_scalled
```

```python
iris.target
```

```python
plt.scatter(pca_scalled[:,0],pca_scalled[:,1],c=iris.target)
```

```python
import scipy.cluster.hierarchy as sc
plt.figure(figsize=(10,7))
dendrogram = sc.dendrogram(sc.linkage(pca_scalled,method='ward'))
plt.title('dendrogram')
sc.dendrogram(sc.linkage(pca_scalled,method='ward'))
plt.title('Dendogram')
plt.xlabel('Sample Index')
plt.ylabel('Eucledian Distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
cluster.fit(pca_scalled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scalled[:,0],pca_scalled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_coefficients = []
for k in range(2,11):
    agglo = AgglomerativeClustering(n_clusters=k, metric='euclidean',linkage='ward')
    agglo.fit(x_scaled)
    score = silhouette_score(x_scaled, agglo.labels_)
    silhouette_coefficients.append(score)
```

```python
silhouette_coefficients
```

```python
plt.plot(range(2,11),silhouette_coefficients)
plt.xticks(range(2,11))
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Coefficient')
plt.show()
```
