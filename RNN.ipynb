```python
df1 = df1.iloc[:,1:2].values
df = df.iloc[:,1:2].values
df1
```

```python
sc = MinMaxScaler()
df1 = sc.fit_transform(df1)
df1
```

```python
df = sc.fit_transform(df)
df
```

```python
x_train = df1[0:1257]
y_train = df1[1:1258]
```

```python
print(f"Before reshaping: {x_train.shape}")
x_train = x_train.reshape(1257,1,1)
print(f"After reshaping: {x_train.shape}")
```

```python
from keras.models import Sequential
from keras.layers import Dense,LSTM
```

```python
model = Sequential()
model.add(LSTM(units = 4, activation= 'sigmoid', input_shape = (None,1)))
```

```python
model.add(Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size =32 , epochs= 200)
```

```python
real_stock_price = df
inputs = real_stock_price
inputs = sc.transform(df)                   # feature scaling
inputs = np.reshape(inputs, (20, 1, 1))           # reshape the data
predicted_stock_price = model.predict(inputs)     # prediction
```

```python
# inverse to the original scale
predicted_stock_price = sc.inverse_transform(predicted_stock_price)

# visualize the result
plt.plot( real_stock_price , color = 'red' , label = 'Real Google Stock Price')
plt.plot( predicted_stock_price , color = 'blue' , label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel( 'time' )
plt.ylabel( 'Google Stock Price' )
plt.legend()
plt.show()
```

```python
training_set = pd.read_csv('/content/Google_Stock_Price_Train.csv')
training_set = training_set.iloc[:, 1:2].values

real_stock_price = training_set
inputs = real_stock_price
inputs = sc.transform(inputs)
inputs = np.reshape(inputs, (1258, 1, 1))
predicted_stock_price = model.predict(inputs)
```
